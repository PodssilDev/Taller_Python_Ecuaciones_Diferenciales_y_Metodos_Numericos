La EDO autónoma:
$$ m\dfrac{dy}{dt}=mg-ky^2 $$
modela la velocidad y(t) de un cuerpo de masa m (en $kg$) que está cayendo bajo la influencia de la gravedad. Debido a que el término $–ky^2$ representa la resistencia del aire, la velocidad de un cuerpo que cae de una gran altura no aumenta sin límite conforme pasa el tiempo t.
* Suponga que el cuerpo tiene masa 2 $kg$ y k = g donde g = 9,8 $m/s^2$

### (a) Construya un campo de direcciones para la EDO autónoma.
* Notemos que los puntos de equilibrio de la ecuacion anterior reemplazada con los datos dados, corresponden a $-\sqrt 2$ y $ \sqrt2$.

```python
# Se importa Matplotlib.pyplot como plt para trabajar con graficos
import matplotlib.pyplot as plt
# Se importa numpy como np para trabajar con arreglos
import numpy as np
```

```python
# Reemplazamos los valores del enunciado para trabajar con la formula dada
# dy/dt = (mg - k^y)/ m
def f(X,Y): return 9.8-4.9*(Y**2) # Definimos la funcion con la que trabajaremos
X,Y=np.meshgrid(np.linspace(-2,2,10), np.linspace(-2,2,10)) # Utilizamos meshgrid para obtener una matriz de puntos
U = 1.0  # Se define la direccion horizontal de las flechas
V = f(X,Y) # Se define la direccion vertical de las flechas
N = np.sqrt(U**2+V**2) # Se prepara para normalizar los vectores U y V
U = U/N # Se normaliza el vector U
V = V/N # Se normaliza el vector N
plt.quiver(X, Y, U, V)  # Se realiza el grafico del campo de direccion
plt.xlabel('t') # Se define el eje X como "t"
plt.ylabel('y') # Se fine el eje Y como "y"
plt.grid() # Se agregan una red al fondo del grafico
```

Como en el gráfico no se ve tan claro los puntos de equilibrio ni sus clasificaciones, se puede realizar un "zoom" a ambos puntos para poder visualizarlos de manera mas clara.

```python
# Nos enfocaremos en el punto de equilibrio 1,4142... utilizando los 
# puntos 1.2 y 1.6 en el eje Y
# Utilizamos meshgrid para obtener una matriz de puntos
X,Y=np.meshgrid(np.linspace(-2,2,10), np.linspace(1.2,1.6,10))
U = 1.0  # Se define la direccion horizontal de las flechas
V = f(X,Y) # Se define la direccion vertical de las flechas
N = np.sqrt(U**2+V**2) # Se prepara para normalizar los vectores U y V
U = U/N # Se normaliza el vector U
V = V/N # Se normaliza el vector N
plt.quiver(X, Y, U, V)  # Se realiza el grafico del campo de direccion
plt.xlabel('t') # Se define el eje X como "t"
plt.ylabel('y') # Se fine el eje Y como "y"
plt.grid() # Se agregan una red al fondo del grafico
```

```python
# Nos enfocaremos en el punto de equilibrio -1,4142... utilizando los
# puntos -1.6 y -1.2 en el eje Y
# Utilizamos meshgrid para obtener una matriz de puntos
X, Y = np.meshgrid(np.linspace(-2, 2, 10), np.linspace(-1.6, -1.2, 10))
U = 1.0   # Se define la direccion horizontal de las flechas
V = f(X, Y)  # Se define la direccion vertical de las flechas
N = np.sqrt(U ** 2 + V ** 2)  # Se prepara para normalizar los vectores U y V
U = U / N  # Se normaliza el vector U
V = V / N  # Se normaliza el vector N
plt.quiver(X, Y, U, V)   # Se realiza el grafico del campo de direccion
plt.xlabel('t')  # Se define el eje X como "t"
plt.ylabel('y')  # Se fine el eje Y como "y"
plt.grid()  # Se agregan una red al fondo del grafico
```

### (b) Determine y clasifique los puntos de equilibrio en atractor, repulsor o nodo.

Primero, hacemos el despeje de $\dfrac{dy}{dt}$
$$ m\dfrac{dy}{dt}=mg-ky^2 $$

$$ \dfrac{dy}{dt}=\frac {mg-ky^2}{m} $$
Reemplazando los datos: m = 2, k = g = 9,8
$$ \dfrac{dy}{dt}=9,8-4,9y^2 $$
Igualamos a 0 para encontrar los puntos de equilibrio
$$ 0 = 9,8-4,9y^2 $$

$$ 4,9y^2 = 9,8 $$

$$ y^2 = \frac{9,8}{4,9} $$

$$ y = \sqrt 2 $$

$$ y = -\sqrt 2 $$

Ahora, guiandonos por los graficos de la parte A, los puntos de equilibrio se clasifican de la siguiente forma:
#### * $ y = \sqrt 2 $: Atractor o sumidero, pues cualquier solucion cercana a $ y = \sqrt 2 $ tiende a $\sqrt 2$  a medida que t aumenta.
#### * $ y = -\sqrt 2 $: Fuente o repulsor, pues cualquier solucion cercana a $ y = -\sqrt 2 $ tiende a $-\sqrt 2$  a medida que t decrece.

### (c) Utilice la información anterior para encontrar la velocidad límite conforme t → ∞.

Como tenemos tres intervalos de $y$ distintos, debemos considerar que la velocidad límite conforme t → ∞ puede ser distinta en esos tres intervalos. Considerando esto, las gráficas realizadas en la parte A y las clasificaciones de la parte B:
#### * $\sqrt 2  < y < ∞ $: La velocidad límite conforme t → ∞ es $\sqrt 2$, pues las flechas cada vez se atraen hacia $ y = \sqrt 2$ a medida que t aumenta.
#### * $-\sqrt 2  < y < \sqrt 2 $: La velocidad límite conforme t → ∞ es $\sqrt 2$, pues las flechas cada vez se atraen hacia $ y = \sqrt 2$ a medida que t aumenta.
#### * $ -∞ < y < -\sqrt 2 $: La velocidad límite conforme t → ∞ es -∞ pues las flecha cada vez se van hacia $y = -∞$ a medida que t aumenta.
